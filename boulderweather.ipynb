```python
import pandas as pd
import numpy as np
import dateutil
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weatherdata.csv")
```

```python
df['DATE'] = df['DATE'].apply(dateutil.parser.parse, dayfirst=True)
df['DATE'] = pd.to_datetime(df['DATE'])
dfi = df.set_index(['DATE'])
```

```python
#Plot temperature by date
df.plot(x="DATE", y="HourlyDryBulbTemperature", title="Temperature over time in Boulder, CO - 2022")
```

```python
#Select data from date ranges
jan = dfi.loc['2022-01']
feb = dfi.loc['2022-02']
mar = dfi.loc['2022-03']
apr = dfi.loc['2022-04']
may = dfi.loc['2022-05']
```

```python
ax1 = mar.plot(kind='scatter', x='HourlyDryBulbTemperature', y='HourlyRelativeHumidity', color='r')    
ax2 = apr.plot(kind='scatter', x='HourlyDryBulbTemperature', y='HourlyRelativeHumidity', color='g', ax=ax1)    
ax3 = may.plot(kind='scatter', x='HourlyDryBulbTemperature', y='HourlyRelativeHumidity', color='b', ax=ax1)

plt.title("Humidity and Temperature by Month in Boulder")
plt.legend(['March', 'April', 'May'])
print(ax1 == ax2 == ax3)
```
